## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO : Exercise 1
# Onofre Liscano
# Importo random library y comento seed para resultados reproducibles
# Importo pandas

import random
import pandas as pd
#random.seed (42)

# E(Î©) = {36}                pues son dos dados y el espacio muestral es 6x6 (n*m)
# A    = {2,4,6,8,10,12}     pues son dos dados, y nos dicen total score (total % 2 == 0)
# B    = {8,9,10,11,12}      pues son dos dados y seria total score mayor a 7 (total > 7)



# Me decidi por usar un Data Frame (DF) para visualizar comodamente los resultados
# Creo mis columnas para el DF

columnas = ["ID Tirada", "ðŸŽ² #1", "ðŸŽ² #2", "Total en la tirada", "Se Cumple?"]
tiradas = pd.DataFrame(columns=columnas)

for i in range(1000):

    # Almanceno los resutlados de cada dado por separado en variables
    resultadoDado1 = (random.randint(1,6))
    resultadoDado2  = (random.randint(1,6))

    # Obtengo el resultado de cada uno y sumo para tener el Total Score
    totalScore = resultadoDado1 + resultadoDado2

    # Evaluo totalScore si cumple con A o B
    condicionA = (totalScore % 2 == 0)
    condicionB = (totalScore > 7)
    exitoCondicion = condicionA or condicionB
    
    # Creo mis tiradas del DF por row usando (i)
    tiradas.loc[i] = {
        "ID Tirada": i,
        "ðŸŽ² #1": resultadoDado1,
        "ðŸŽ² #2": resultadoDado2,
        "Total en la tirada": totalScore,
        "Se Cumple?": exitoCondicion 
    }

print (tiradas.head())
probabilidadEstimada = tiradas["Se Cumple?"].mean()
print (f"\nProbabilidad estimada en 1.000 intentos: {probabilidadEstimada}")

# Los Ãºnicos valores permitidos cuando Se Cumple? es False
esperado = {3, 5, 7}
observado = set(tiradas[~tiradas["Se Cumple?"]]["Total en la tirada"].unique())

if observado == esperado:
    print("Sanity check OK: solo 3, 5 y 7 fallaron.")
else:
    print("Algo anda mal, aparecieron:", observado - esperado)
  
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
```

```python
# TODO : Exercise 2
# Onofre Liscano
# Importo random library y pandas

import random
import pandas as pd

# E(Î©) = {(W),(R),(G)}^5                          pues son 10/60 (WHITE), 20/60 (RED), 30/60 (GREEN)
# A    = {3(W),2(R),0(G)}                         pues tres blancas, 2 rojas y cero verdes
# B    = {[W,W,W,W,W],Â [R,R,R,R,R],Â [G,G,G,G,G]}  pues son todas las mismas    


# Nota importante:
# A âˆ© B = âˆ…   (vacÃ­o), porque no se puede tener a la vez 3W2R y tambiÃ©n 5 iguales.
# Por tanto, A y B son eventos MUTUAMENTE EXCLUYENTES.
#
# Entonces:
# P(A âˆª B) = P(A) + P(B)
# P(A âˆ© B) = 0
# Aâ€² = todas las secuencias que no sean 3W2R

# CREAMOS LAS BOLAS 60 en total
# 0  al 9  â†’ WHITE
# 10 al 30 â†’ RED
# 31 al 60 â†’ GREEN

ball_box = {}
for i in range(60):
    if i < 10:
        ball_box[i] = "WHITE"
    elif (i < 30):
        ball_box[i] = "RED"
    else:
        ball_box[i] = "GREEN"
   
# Funcion para tomar una bola al azar
def getAball(indice):
     return ball_box[indice]
    
#  Funcion para tomar todas las bolas del box
def getAllBalls():
     return ball_box 


# Imprimimos una bola al azar
indice = random.randint(0, 59)
print (f"\nLa bola tomada al azar es: \n#{indice}-{getAball(indice)}")

# Imprimos todo el box
print (f"\nTodo el set de bolas es: \n{getAllBalls()}")

# Hago un Data Frame (DF) para visualizar comodamente los resultados
# Creo mis columnas para el DF

columnas = ["ID Experimento", "âšª WHITE", "ðŸ”´ RED", "ðŸŸ¢ GREEN", "Se Cumple?"]
experimento = pd.DataFrame(columns=columnas)

# Creo un 1er for para el numero total de intentos (1.000)
for i in range(1000):  
    white = 0
    red = 0
    green = 0
    # Por cada uno de los 1.000 intentos voy a sacar 5 bolas
    # por eso creo este segundo for
    for j in range(5):
        indice = random.randint(0,59)
        color = getAball(indice)

        if color == "WHITE":
            white += 1
        elif color == "RED":
            red += 1
        else:
            green += 1

    condicionA = (white == 3 and red == 2)
    condicionB = (white == 5 or red == 5 or green == 5)
    exitoCondicion = condicionA or condicionB
                
    experimento.loc[i] = {
        "ID Experimento": i,
        "âšª WHITE": white,
        "ðŸ”´ RED": red,
        "ðŸŸ¢ GREEN": green,
        "Se Cumple?": exitoCondicion 
    }

print (experimento.head())
probabilidadA = ((experimento["âšª WHITE"] == 3) & (experimento["ðŸ”´ RED"] == 2)).mean()
probabilidadB = ( (experimento["âšª WHITE"] == 5) | (experimento["ðŸ”´ RED"] == 5) | (experimento["ðŸŸ¢ GREEN"] == 5) ).mean()

print (f"\nProbabilidad del evento A en 1.000 intentos: {probabilidadA}")
print (f"\nProbabilidad del evento B en 1.000 intentos: {probabilidadB}")
```
